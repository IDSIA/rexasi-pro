```python
import pandas as pd
import glob
```

```python
experiments = [str(i).zfill(2) for i in range(15)]
cameras = ['drone', 'wheelchair']
models = {'v5':'YOLO v5 small', 'FT':'YOLO v8 nano'}
suffix = ['v5', 'FT']
limits_w = [190, 263, 206, 205, 167, 260, 367, 237, 188, 269, 251, 233, 304, 454, 190]
limits_d = [190, 262, 203, 208, 166, 260, 367, 237, 189, 273, 254, 235, 307, 455, 190]
```

```python
def apply_formula(cell):
    if pd.notnull(cell):
        return (0.24 * 1000) / cell
    else:
        return cell
```

```python
dfs = []
for i, e in enumerate(experiments):
    for cam in cameras:
        for s in suffix:
            df = pd.read_csv(f'{e}_{cam}_{s}.csv', index_col=0)
            df = df.applymap(apply_formula)
            if cam == 'drone':
                df_cut = df.iloc[:limits_d[i]].copy()  # Ensure a copy is made
            elif cam == 'wheelchair':
                df_cut = df.iloc[:limits_w[i]].copy()  # Ensure a copy is made

            df_cut.reset_index(inplace=True)
            df_cut['Experiment'] = i
            df_cut['Camera'] = cam
            df_cut['Model'] = models[s]
            print(df_cut.columns)
            dfs.append(df_cut)

            #custom_order = ['Experiment', 'Camera', 'Model', 'Filter',]  
            #df_sorted = df_cut.reindex(columns=custom_order)
            #df_sorted.to_csv(f'final/{e}_{cam}_{s}.csv')
```

```python
csv_files = glob.glob('final/*.csv')
csv_files
```

```python
# Read each CSV file into a list of dataframes
#dfs = [pd.read_csv(file, index_col=0) for file in csv_files]

# Concatenate all dataframes into a single dataframe
df = pd.concat(dfs, ignore_index=True)

# Write the merged dataframe to a single CSV file
df.to_csv('merged_data.csv', index=False)

print("Merged CSV file created successfully.")
```
